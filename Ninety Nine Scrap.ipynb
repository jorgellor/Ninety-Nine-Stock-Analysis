```python
import scrapy

from scrapy.crawler import CrawlerProcess

class NinetyNineSpider(scrapy.Item):
    name = scrapy.Field()


class SpiderNinetyNineSpider(scrapy.Spider):
    name = 'spider_200406'
    start_urls = ['https://ninetynine.com/acciones/']
    BASE_URL = 'https://ninetynine.com/acciones/'

    def parse(self, response):
            absolute_url = self.BASE_URL
            yield scrapy.Request(absolute_url, callback=self.parse_info)

    def parse_info(self, response):
        item = NinetyNineSpider()
        item["name"] = response.xpath('//span[contains(@class,"company__symbol")]/text()').extract()
        print(item["name"])
        return item
                       

# Run the Spider
process = CrawlerProcess()
process.crawl(SpiderNinetyNineSpider)
process.start()
        

print('end')
```
